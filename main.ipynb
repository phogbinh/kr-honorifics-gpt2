# Set parameters

```python
MODEL_NAME = "skt/kogpt2-base-v2"
MODEL_MAX_LENGTH = 128
MODEL_OUTPUT_DIR = "trained_model/"
MODEL_STEPS_NUM = 3000
MODEL_LEARNING_RATE = 1e-3
MODEL_BATCH_SIZE = 2
```

```python
import os
import torch

os.environ["CUDA_VISIBLE_DEVICES"] = "1"
torch.cuda.is_available()
```

# Train

```python
from aitextgen import aitextgen

ai = aitextgen(model=MODEL_NAME, to_gpu=True)
```

```python
train_and_save_model = True
if train_and_save_model:
  ai.train(
    "train_eval.csv",
    line_by_line=True,
    from_cache=False,
    output_dir=MODEL_OUTPUT_DIR,
    num_steps=MODEL_STEPS_NUM,
    learning_rate=MODEL_LEARNING_RATE,
    batch_size=MODEL_BATCH_SIZE
  )
```

# Evaluate

```python
ai = aitextgen(model_folder=MODEL_OUTPUT_DIR, to_gpu=True)
```

```python
ai.generate(prompt="저기 내일 반품할 노트북이 삼성이니 엘지니", max_length=MODEL_MAX_LENGTH)
```
